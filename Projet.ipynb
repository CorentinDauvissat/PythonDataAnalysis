## Data visualization

```python
correlation = X_test.corr()
```

```python
correl = pd.DataFrame(correlation.stack().reset_index().rename(columns={'level_0': 'feature_1', 'level_1': 'feature_2', 0:'corr'}))
correl.head()
```

```python
correl["abs_corr"] = abs(correl["corr"])
correl.head()
```

```python
correl = correl[(correl["feature_1"] != correl["feature_2"]) & (correl["abs_corr"] > 0.8)]
correl.sort_values("abs_corr", ascending = False)
```

```python
#On regarde le nombre de fois où chaque feature a un score de corrélation supérieur à 0.8 avec une autre feature.
#On en déduit les features les plus significatives
uni_correl = correl["feature_1"].unique()
freq = correl.groupby('feature_1').count()["corr"].sort_values(ascending=False)[:45] 
#On ne garde que les 45 premières valeurs pour faire de la visualisation
```

```python
freq
#Les 5 premières features ont un score beaucoup plus gros car plusieurs colonnes portent leur nom. 
#Je fais le choix de les retirer
```

```python
select_feat = list(freq.index)[5:]
select_feat
```

```python
#On créé une nouvelle dataframe avec les 20 features les plus significatives
neoX_train = X_train[select_feat].copy()
neoX_test = X_test[select_feat].copy()
neoX_train
```

```python
sns.heatmap(neoX_train.corr())
```

## Machine learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Régression logistique

```python
logmodel = LogisticRegression(max_iter = 1000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
print(classification_report(y_test,predictions))
```

### Random forest

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

La régression logistique a une meilleure précision que le random forest
